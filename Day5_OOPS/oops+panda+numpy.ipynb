```python
class Counter:
    def __init__(self, count_value):
        self.count_vlaue = count_value

    def increment(self):
        self.count_vlaue += 1

    def get_count(self):
        return self.count_vlaue 

counter1 = Counter(500)
counter2 = Counter(1000)

counter1.increment()
counter2.increment()
print(counter2.get_count())
print(counter1.get_count())
```

```python
class Animal:
    def __init__(self, eat, name):
        self.name = name
        self.eat = eat

    def eat(self):
            print(f"{self.name} is eating")

    def sleep(self):
            print(f"{self.name} is sleeping")

class Dog(Animal):
    def bark(self):
        print(f"Dog saya {self.name}")

class Cat(Animal):
    def meow(self):
        print(f"Cat say meow{self.name}")    

animal_behaviour = Dog(32, "Dogesh Bhai")
animal_behaviour2 = Cat(12, "pinky")

animal_behaviour.bark()
animal_behaviour2.meow()
animal_behaviour.sleep()
```

```python
from functools import reduce


number = [1, 2, 3, 4, 5, 6, 7, 8]

#Map square every number
squared_number = list(map(lambda x : x**2, number))
print(squared_number)

# Filter: keep even numbers
Filtered_numbers = list(filter(lambda x : x % 2 == 0, number))
print(Filtered_numbers)

# Reduce: product of all numbers
reduced_numbers = reduce(lambda x, y : x * y, number)
reduced_numbers
```

```python
data = [10, 12, 11, 13, 12, 14, 15, 100]

df = pd.DataFrame({"Value":data})
df
```

## Formula for IQR
## IQR = Q3 − Q1

## Lower = Q1 − 1.5*IQR
## Upper = Q3 + 1.5*IQR

```python
Q1 = df["Value"].quantile(0.25)
Q3 = df["Value"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliner = df[(df["Value"] < lower) | (df["Value"] > upper)]
print(outliner)
```
