```python
batch = [
    ([1.2, 3.4, 5.6], 0),
    ([2.1, 4.3, 6.5], 1),
    ([3.2, 5.4, 7.6], 0)
]

for features, Label in batch:
    print(features, Label)
    f1, f2, f3 = features
    print(f1, f2, f2)
    a = Label
    print(a)
    print(f1)
    print(f2)
    print(f3)
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(f"Union: {A.union(B)}")
```

```python
A = {1, 2, 3, 4}
B = {3, 4}
print(f"subset:{B.issubset(A)}")
print(f"intersection: {(A.issuperset(B))}")
```

```python
print("=== Example 1: Building Vocabulary ===")
documents = [
    "machine learning is awesome",
    "deep learning is powerful",
    "machine learning and deep learning"
]

vocabulary = set()
for doc in documents:
    words = doc.lower().split()
    vocabulary.update(words)
    print(f"Updated vocabulary: {vocabulary}")
```
