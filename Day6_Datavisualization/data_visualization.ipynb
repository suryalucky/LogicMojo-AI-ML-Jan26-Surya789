```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd

sns.set_theme(style="whitegrid")
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y, label="Sine Wave")
plt.title("Simple Line Plot Using Matplotlib")
plt.xlabel("X-axis (Time/index)")
plt.ylabel("Y-axis (Amplitude)")
plt.legend()
plt.show()
```

## Seanborn Example

```python
tips = sns.load_dataset("tips")
tips
```

## Distribution Chart

```python
sns.histplot(data=tips, x="total_bill", kde=True, bins=55, hue="time")
```

### Skewness --- Right skewed curve

```python
tips["time"].value_counts()
```

### Box plot

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
```

```python
tips[(tips["day"] == "Sat") & (tips["smoker"] == "Yes")]
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.jointplot(data=tips, x = 'total_bill', y = 'tip', kind='hex')
```

### Matrix Plot (Correlations)

```python
matrix_col = tips.select_dtypes(include=(np.number))
corr = matrix_col.corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
```

### Pie and Donut Charts

```python
day_counts = tips['day'].value_counts()
day_counts
```

```python
plt.pie(day_counts, labels=day_counts.index, autopct="%1.1f%%")
```

```python
plt.pie(day_counts, labels=day_counts.index, autopct="%1.1f%%", pctdistance=0.85)
center_circle = plt.Circle((0,0), 0.5, fc='white')
fig = plt.gcf() # gcf means Get current figure 
fig.gca().add_artist(center_circle) # Add circle to axis, gca - get current axis

```

### Bar char

```python
sns.barplot(data=tips, x="day", y="total_bill")
```
