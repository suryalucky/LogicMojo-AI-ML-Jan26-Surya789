```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import _california_housing

housing = _california_housing.fetch_california_housing(as_frame = True)

x = housing.data
y = housing.target
```

```python
print(type(housing))
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
print(x.head())
print(y.head())
```

```python
print(x.columns)
```

```python
print(x.iloc[0])
print(y.iloc[0])
```

```python
print(x.info())
print(x.isnull().sum())
print(y.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
coef_df = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': lr.coef_
})

print(coef_df)
```

```python
y_predit = lr.predict(X_test)

print(y_predit[:5])
print(y_test[:5])
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
mae = mean_absolute_error(y_test, y_predit)
mse = mean_squared_error(y_test, y_predit)
r2 = r2_score(y_test, y_predit)
rmse = np.sqrt(mse)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'R2 Score: {r2}')
print(f'RMSE: {rmse}')
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes

data = load_diabetes(as_frame=True)
x = data.data.copy()
y = data.target.copy()
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
print(x.head())
print(y.head())
```

```python
print(x.columns)
print(x.info())
print(x.describe())
print(x.isnull().sum())
print(y.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
mean_guess = y_train.mean()
baseline_predictions = np.full(shape=y_test.shape, fill_value=mean_guess)
baseline_mae = mean_absolute_error(y_test, baseline_predictions)
print(f'Baseline MAE: {baseline_mae}')
print("Mean guess:", mean_guess)
```

```python
line_reg = LinearRegression()
line_reg.fit(X_train, y_train)
```

```python
y_predit = line_reg.predict(X_test)

print(y_predit[:5])
print(y_test.iloc[:5])
```

```python
mae = mean_absolute_error(y_test, y_predit)
mse = mean_squared_error(y_test, y_predit)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_predit)

print("Baseline MAE:", baseline_mae)
print("Model MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2:", r2)
```
