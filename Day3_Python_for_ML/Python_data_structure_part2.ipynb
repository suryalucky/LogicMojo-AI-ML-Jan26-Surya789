```python
batch = [
    ([1.2, 3.4, 5.6], 0),
    ([2.1, 4.3, 6.5], 1),
    ([3.2, 5.4, 7.6], 0)
]

for features, Label in batch:
    print(features, Label)
    f1, f2, f3 = features
    print(f1, f2, f2)
    a = Label
    print(a)
    print(f1)
    print(f2)
    print(f3)
```

```python
print("=== Example 1: Building Vocabulary ===")
documents = [
    "machine learning is awesome",
    "deep learning is powerful",
    "machine learning and deep learning"
]

vocabulary = set()
for doc in documents:
    words = doc.lower().split()
    vocabulary.update(words)
    print(f"Updated vocabulary: {vocabulary}")
```

```python
# Dictionary methods
Score = {'Alice': 90, 'Bob': 85, 'Charlie': 92}
print(Score.keys())
print(Score.values())
print(Score.items())
```

```python
for name in Score:
    print(f"{name}: {Score[name]}")
```

```python
for student, score in Score.items():
    print(f"{student}: {score}")
```

```python
new_score = {'David': 88}
Score.update(new_score)
print(Score)
```

```python
# word fequency count
text = "data science is an interdisciplinary field that uses scientific methods processes algorithms and systems to extract knowledge and insights from structured and unstructured data"
word_freq = {}
for word in text.split():
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1
print(word_freq) 
```

```python
from collections import Counter
word_freq = Counter(text.split())
print(word_freq)
```

```python
Score = {'Alice': 90, 'Bob': 85, 'Charlie': 92}
Passing_Score = {name: score for name, score in Score.items() if score >= 90}
print(Passing_Score)
```

```python
point = (2, 3)
rgb = (225, 0, 0)
single = (42,) # Typle with a single element, if there is no comma it will be considered as integer
empty = () # empty tuple
no_params = 1, 2, 3, 5 # Typle packing without parentheses

print(point)
print(rgb)
print(single)
print(empty)
print(no_params)
```

```python
#Accesssing tuple elements
print(point[0])
print(rgb[-1])
print(rgb[1:3])
```

```python
#Common operations on tuples
print(len(point))
print(max(rgb))
print(min(rgb))
print(rgb.count(225))
print(rgb.index(0))
print(point + (4, 5))
print(point * 2)
```
